### Mining Data 

```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome("/Users/m/Downloads/chromedriver-3")
driver.get("http://romainvaleri.online.fr/")

#Find the button to generate text: 

button = driver.find_element_by_id("b_generer")

#Click like jaja
i=0
while i < 1500:    
    button.click()
    i=i+1



#Extract Data : 

z = 1
DB_Sentence = []
while z < i:
    phrase_id = "phrase_"+str(z)
    Sentence = driver.find_element_by_id(phrase_id).text
    DB_Sentence.append(Sentence)
    z=z+1

```

```python
import pandas as pd 

df= pd.DataFrame(list(DB_Sentence), columns = ["Sentence"])
df["Eval 1"] = 0
df["Eval 2"] = 0
df["Eval 3"] = 0
```

### CSV for the Deliverable

```python
#Change parameters if you want

df.to_csv(r'/Users/m/Downloads/export_dataframe.csv')
```
